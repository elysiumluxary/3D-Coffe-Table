```python
# 1. İnterfeys üçün lazım olan kitabxanaları quraşdırırıq
print("Professional SaaS İnterfeysi (Gradio) yüklənir...")
!pip install -q gradio requests

import gradio as gr
import requests
import json

# 2. Ollama Llama 3 ilə əlaqə quran arka plan funksiyası
def evelysium_brain(user_message, system_prompt, chat_history):
    url = "http://localhost:11434/api/generate"

    # Keçmiş dialoqu və yeni mesajı Sokratik arxitekturaya uyğun birləşdiririk
    full_prompt = f"System: {system_prompt}\n\n"
    for user_h, bot_h in chat_history:
        full_prompt += f"User: {user_h}\nAI: {bot_h}\n"
    full_prompt += f"User: {user_message}\nAI:"

    payload = {
        "model": "llama3",
        "prompt": full_prompt,
        "stream": False
    }

    try:
        response = requests.post(url, json=payload)
        response_json = json.loads(response.text)
        bot_response = response_json.get("response", "Xəta: Cavab alınmadı.")

        # Tarixçəni yeniləyirik
        chat_history.append((user_message, bot_response))
        return "", chat_history
    except Exception as e:
        chat_history.append((user_message, f"Sistem xətası: {str(e)}"))
        return "", chat_history

# 3. Mükəmməl UI Dizaynı (Evrensel İnsan Zihni Platformu Prototipi)
with gr.Blocks(theme=gr.themes.Soft()) as elysium_ui:
    gr.Markdown("# 🧠 Evrensel İnsan Zihni Platformu — ELYSIUM v1.0")
    gr.Markdown("### *Sokratik Diyalektik, Psikolojik Manipülasyon ve Güç Dinamikleri Analiz Laboratuvarı*")

    with gr.Row():
        # Sol Panel: Strateji və Zəka Ayarları
        with gr.Column(scale=1):
            gr.Markdown("### ⚙️ Zəka Katmanı Ayarları")
            sys_input = gr.Textbox(
                label="AI-ın Psixoloji Rolu (System Prompt)",
                lines=10,
                value="Sen Sokratik diyalektik ve Transaksiyonel Analiz (TA) uzmanı bir Baş Filozofsun. Karşındaki insanın cümlelerindeki mantık hatalarını, safsataları ve psikolojik manipülasyonları (Kurban-Kurtarıcı-Zalim üçgeni) doğrudan söyleme. Ona öyle Sokratik sorular sor ki, kendi manipülasyonunu ve zafiyetini kendisi fark etsin. Derin, analitik ve sarsıcı ol."
            )
            gr.Markdown("> **Not:** Bu paneldən AI-ın insan zəhninə hansı nöqtədən hücum edəcəyini dəyişə bilərsən.")

        # Sağ Panel: Canlı Sokratik Dialoq Ekranı
        with gr.Column(scale=2):
            gr.Markdown("### 💬 İnsan Zihni Analiz Ekranı")
            chatbot = gr.Chatbot(label="Sokratik Dialoq Tarixçəsi")
            msg = gr.Textbox(label="İnsanın Cümləsi (Analiz ediləcək mətn)", placeholder="Bura bir cümlə yazın və Enter sıxın...")
            clear = gr.Button("Söhbəti Sıfırla")

    # Düymələrin və daxiletmənin funksiyalara bağlanması
    msg.submit(evelysium_brain, [msg, sys_input, chatbot], [msg, chatbot])
    clear.click(lambda: None, None, chatbot, queue=False)

# 4. İnterfeysi Dünyaya Açırıq (SHARE=TRUE 100% ŞƏRTDİR)
print("\nElysium Portalı Dünyaya Açılır...")
elysium_ui.launch(share=True, debug=True)
```

```python
print("Hibrid Zəka Arxitekturası Qurulur (Llama 3 + Gemini)...")
!pip install -q google-generativeai gradio requests

import gradio as gr
import google.generativeai as genai
import requests
import json

# Əsas Zəka Mühərriki Funksiyası
def elysium_hybrid_engine(user_message, model_choice, system_prompt, custom_gemini_key, chat_history):

    # ---- 1. SEÇİM: GOOGLE GEMINI ----
    if model_choice == "Google Gemini 1.5 Flash (Bulud/Sürətli)":
        if not custom_gemini_key or custom_gemini_key.strip() == "":
            chat_history.append((user_message, "⚠️ Xəta: Gemini modelini seçdiniz, lakin API Key daxil etmədiniz! Zəhmət olmasa sol paneldən API Key yazın və ya Llama 3 modelinə keçin."))
            return "", chat_history

        # Gemini-ni daxil edilən açarla oyadırıq
        genai.configure(api_key=custom_gemini_key.strip())
        model = genai.GenerativeModel(model_name="gemini-1.5-flash", system_instruction=system_prompt)

        gemini_history = []
        for user_h, bot_h in chat_history:
            gemini_history.append({"role": "user", "parts": [user_h]})
            gemini_history.append({"role": "model", "parts": [bot_h]})

        chat = model.start_chat(history=gemini_history)
        try:
            response = chat.send_message(user_message)
            chat_history.append((user_message, response.text))
            return "", chat_history
        except Exception as e:
            chat_history.append((user_message, f"Google Bulud Xətası: {str(e)}"))
            return "", chat_history

    # ---- 2. SEÇİM: YERLİ LLAMA 3 (PULSUZ/LİMİTSİZ) ----
    else:
        url = "http://localhost:11434/api/generate"
        full_prompt = f"System: {system_prompt}\n\n"
        for user_h, bot_h in chat_history:
            full_prompt += f"User: {user_h}\nAI: {bot_h}\n"
        full_prompt += f"User: {user_message}\nAI:"

        payload = {"model": "llama3", "prompt": full_prompt, "stream": False}
        try:
            response = requests.post(url, json=payload)
            response_json = json.loads(response.text)
            bot_response = response_json.get("response", "Xəta: Llama modelindən cavab alınmadı.")
            chat_history.append((user_message, bot_response))
            return "", chat_history
        except Exception as e:
            chat_history.append((user_message, f"Yerli Server (Ollama) Xətası! Arxa planda Ollama-nın aktiv olduğundan əmin olun. Error: {str(e)}"))
            return "", chat_history

# 🖥️ MÜKƏMMƏL HİBRİD UI DİZAYNI
with gr.Blocks(theme=gr.themes.Default()) as elysium_v3:
    gr.Markdown("# 🧠 ELYSIUM v3.0 — Hybrid Human Mind Platform")
    gr.Markdown("### *İstifadəçiyə Seçim Azadlığı Verən Sokratik və Psixoloji Analiz Laboratoriyası*")

    with gr.Row():
        # SOL PANEL: İdarəetmə və Model Seçimi
        with gr.Column(scale=1):
            gr.Markdown("### ⚙️ Arxitektura Tənzimləmələri")

            # Əsas Strateji Keçid Qutusu (Dropdown)
            model_select = gr.Dropdown(
                choices=["Meta Llama 3 (Yerli/Pulsuz/Limitsiz)", "Google Gemini 1.5 Flash (Bulud/Sürətli)"],
                value="Meta Llama 3 (Yerli/Pulsuz/Limitsiz)",
                label="İstifadə Olunacaq AI Mühərriki"
            )

            gemini_key_input = gr.Textbox(
                label="Google Gemini API Key (Seçimə bağlıdır)",
                placeholder="AIzaSy...",
                type="password"
            )

            sys_prompt_input = gr.Textbox(
                label="AI-ın Psixoloji Təlimatı (System Prompt)",
                lines=8,
                value="Sen Sokratik diyalektik ve Transaksiyonel Analiz uzmanı bir Baş Filozofsun. Karşındakı insanın mantık hatalarını yüzüne vurma, ona Sokratik sorular sorarak kendi çelişkilerini kendisinin görmesini sağla. Sarsıcı ve derin ol."
            )
            gr.Markdown("> **Proje Yöneticisi Qeydi:** Gündəlik testlər üçün Llama-nı seçin. Sürət və dərin Azərbaycan dili analizi üçün Gemini-yə keçib öz API açarınızı daxil edin.")

        # SAĞ PANEL: Canlı Dialoq
        with gr.Column(scale=2):
            gr.Markdown("### 💬 Sokratik Eksperiment Meydançası")
            chatbot_display = gr.Chatbot(label="Dialoq Tarixçəsi")
            user_input = gr.Textbox(label="İnsanın Cümləsi", placeholder="Mətninizi yazın və Enter düyməsini sıxın...")
            clear_btn = gr.Button("Söhbəti Təmizlə")

    # Bağlantılar
    user_input.submit(
        elysium_hybrid_engine,
        [user_input, model_select, sys_prompt_input, gemini_key_input, chatbot_display],
        [user_input, chatbot_display]
    )
    clear_btn.click(lambda: None, None, chatbot_display, queue=False)

print("\nElysium v3.0 Portalı Canlı Olaraq Dünyaya Açılır...")
elysium_v3.launch(share=True, debug=True)
```
